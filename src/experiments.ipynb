```python
def generate_problems(per_class = 2):
    problems = []
    max_total_weights = [10, 20, 30]
    item_counts = [5, 10, 20]
    test_cases = []
    for iteration in range(per_class):
        for total_weight in max_total_weights:
            for count in item_counts:
            
                test_case = {
                    'maximum_total_weight': total_weight,
                    'item_count': count
                }
                test_cases.append(test_case)
                
    for case in test_cases:
        problems.append(knapsack(minimum_item_weight=0,maximum_item_weight=10, **case))
    return problems
```

```python
problems = generate_problems()
print(len(problems))
```

```python
algs = []
pop_sizes = [6, 10, 20]
crossover_types = ["one point", "uniform", "two point"]
mutation_probabilities = [0.05, 0.1, 0.2]
selection_methods = ["roulette", "tournament"]
roulette_params = [0,0.2,0.5]
tournament_fracs = [0.2,0.4,0.6]
for problem in problems:
    for pop_size in pop_sizes:
        for crossover_type in crossover_types:
            for mutation_probability in mutation_probabilities:
                for selection_method in selection_methods:
                    if selection_method == 'tournament':
                        for tournament_frac in tournament_fracs:
                            for tournament_frac in tournament_fracs:
                                alg = genetic_alg(
                                    pop_size=pop_size,
                                    genome_size=problem.get_item_count(),
                                    crossover_type=crossover_type, 
                                    mutation_method="gene bitflip", 
                                    mutation_probability=mutation_probability,
                                    problem=problem,
                                    genome_type='bitstring',
                                    selection_method=selection_method, 
                                    tournament_frac=tournament_frac,tournament_prob=tournament_frac)
                                algs.append(alg)
                    else:
                        for roulette_param in roulette_params:
                            alg = genetic_alg(
                                    pop_size=pop_size,
                                    genome_size=problem.get_item_count(),
                                    crossover_type=crossover_type, 
                                    mutation_method="gene bitflip", 
                                    mutation_probability=mutation_probability,
                                    problem=problem,
                                    genome_type='bitstring',
                                    selection_method=selection_method, 
                                    roulette_param=roulette_param)
                            algs.append(alg)
```

```python
from tqdm import tqdm
for alg in tqdm(algs):
    alg.run(50)
```

```python
import pickle
def save_object(obj, filename):
    with open(filename, 'wb') as outp:  # Overwrites any existing file.
        pickle.dump(obj, outp, pickle.HIGHEST_PROTOCOL)

# sample usage
for i, alg in enumerate(algs):
    save_object(algs[i], 'objects1/'+str(i)+'.pkl')
```
